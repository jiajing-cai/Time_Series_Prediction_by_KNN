# Build KNN Models

# two dimensions

```python
df.loc[:,[18,19,20]]
```

```python
#split training and test set

X_train = df.loc[1:7,[18,19]]
y_train = df.loc[1:7,[20]]

X_test = df.loc[8:8,[18,19]]
y_test = df.loc[8:8,[20]]
```

```python
#KNN Regression
k = [2,3,4,5,6,7]
def knn_model(x_train, y_train, x_test, y_test, k):
    for i in k:
        knn = KNeighborsRegressor(n_neighbors = i)
        knn.fit(x_train, y_train)
        #generate enough predict data
        y_pred = knn.predict(x_test)
        y_test['when k='+ str(i)] = y_pred
    return y_test
```

```python
print(knn_model(X_train, y_train, X_test, y_test, k))
```
